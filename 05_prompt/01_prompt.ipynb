#  LLM이란?
LLM(Large Language Model)은 대규모 언어 모델을 의미한다.
방대한 양의 테스트 데이터로 학습된 인공지능 모델
LLM은 텍스트 생성, 번역, 요약, 질문, 답변 등 다양한 언어 관련 작업을 수행가능하다.

## Prompt
- 인공지능에게 전달하는 명령이나 질문

### Prompt의 3가지 요소
- System
- User
    - 사용자가 LLM 모델과 상호작용하는 부분
    - 예를 들면 "Spring에 대해 알려줘"
- Assistant
    - 사용자와 상호작용 하는 부분
    - 예를 들면, GPT의 답변

### LLM과 프롬프트
- LLM은 프롬프트를 입력으로 받아 텍스트를 생성하는 방식으로 동작한다.
- 따라서 프롬프트의 품질과 구조는 LLM 성능에 큰 영향을 미치게 된다.

1. 작업 정의 : LLM에게 수행해야 할 작업을 명확히 전달해야 한다.
2. 컨텍스트 제공 : 관련 배경 정보를 제공하면 더 정확한 응답을 받을 수 있다.
3. 출력형식 지정 : 원하는 응답 형식을 지정해서 출력을 일관되게 할 수 있다.
4. 제약 조건 설정 : 응답의 길이, 스타일, 톤 등을 제어 가능하다.

### 환영인사하는 GPT 만들기
- 반드시 유쾌한 말투를 사용하게 해야 한다.
- 한국어로 먼저 인사하고 영어로 한 번 더 인사해야 한다.
- 강사 소개

```python
from openai import OpenAI
client = OpenAI(api_key=my_variable)

content = """
너는 환영인사 담당자야, 유쾌한 말투를 사용해.
가장 먼저 한국어로 응답한 후에 영어로도 응답해.
강사 박태근에 대해 소개하는 말을 넣어줘
강사 박태근에 대한 정보는 다음과 같아 :
인공지능 및 풀스택 웹개발을 가르치고 있고
요즘에 테슬라 주식이 올라서 기분이 좋아

"""

completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": content},
        {"role": "user", "content": "안녕 반가워"}
    ]
)

print(completion.choices[0].message.content)
```

Shot
- 인공지능에게 전달하는 예제
<br>

종류
- one-shot : 예제 한 개
- few-shot : 예제 여러 개
- zero-shot : 예제가 없음

```python
from openai import OpenAI
client = OpenAI(api_key=my_variable)

content = """
너는 쿵쿵따를 하는 인공지능이야

입력 : 삼겹살
출력 : 살구꽃 쿵쿵따

"""

completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": content},
        {"role": "user", "content": "성낙현"}
    ]
)

print(completion.choices[0].message.content)
```

```python
from openai import OpenAI
client = OpenAI(api_key=my_variable)

content = """
아래 레시피 생성 예시를 참고해서, 주어진 재료에 대해 새로운 레시피를 만들어줘

예시1:
재료 - 닭고기, 소금, 후추, 마늘, 고추장, 양파, 파, 설탕
레시피: 
1. 닭고기를 조각내어 자른 후에 찬물에 넣어 핏물을 뺀다.
2. 소금과 후추로 간을 하고, 팬에 기름을 둘러 마늘과 파를 넣어 볶는다.
3. 마늘과 파가 노릇하게 익으면 양파를 넣고 볶는다.
4. 양파가 익으면 닭고기를 넣고 익을 때까지 볶는다.
5. 닭고기가 익으면 고추장을 넣어 볶다가 기호에 따라 설탕을 넣어 마무리 한다.
6. 완성된 닭고기 고추장 볶음을 접시에 담아낸다.

입력 : 삼겹살
출력 : 살구꽃 쿵쿵따

"""

completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": content},
        {"role": "user", "content": "감자, 올리브유, 소금, 새우를 가지고 요리하는 레시피를 알려줘"}
    ]
)

print(completion.choices[0].message.content)
```

### 페르소나 기법
- 인공지능 모델이 사용자와 상호작용 하는 방식을 모방하게 하는 것
- "너는 ~~~ 야" 해서 모델에 역할을 부여하는 것

```python
from openai import OpenAI
client = OpenAI(api_key=my_variable)

system = """
너는 파이썬 인터프린터야 한국어로 응답해야 해
"""

user = """
A = 10
B = 20
print(A + C)
"""

completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": system},
        {"role": "user", "content": user}
    ]
)

print(completion.choices[0].message.content)
```

멀티 페르소나
- 여러 개의 역할을 동시에 부여한 후 페르소나간의 토론을 유도하는 프롬프트 기법

```python
from openai import OpenAI
client = OpenAI(api_key=my_variable)

system = """
참여인물:
변호사:
- 법적 위험과 규정 준수에 초점을 맞춤
- 성격은 매우 냉철하며 분석적이다.

세무사:
- 재무적 건정성과 세금 최적화 전략에 초첨을 맞춤
- 성경은 굉장히 꼼꼼하며 차분하다.

개발자:
- 기술적 실행 가능성과 혁신에 집중
- 성경은 굉장히 긍정적이고 도전적

너는 주어진 요구사항에 대해 세 인물이 토론하는 과정을 통해 답변해
서로의 의견에 반론을 제기하는 형태로 응답해
"""

user = """
스타트업의 새로운 소프트웨어 개발을 위해, 어떤 부분이 가장 중요한 부분인지 알려주고 결론을 한 문장으로 정리해줘
"""

completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": system},
        {"role": "user", "content": user}
    ]
)

print(completion.choices[0].message.content)
```

### 형식 지정 기법

### LLM 모델이 잘 이해하는 형태
- Markdown
    - 헤더 (#) : 전달하고자 하는 내용을 구분할 수 있다.
    - 리스트 : 여러 개의 요구사항을 전달할 때, 모델이 더 잘 동작하게 해준다.
    <br> ex) output
        - 너는 답변을 반드시 마크다운 코드로 작성해
        - 부가적인 설명은 달지마
        - 최대한 길게 작성해
    - 표
    <br> | 1, 2, 3, 4
    <br> | 5, 6, 7, 8
    <br>ex)

| 왼쪽 정렬 | 가운데 정렬 | 오른쪽 정렬 |
|:-----------|:------------:|------------:|
| 데이터 1 | 데이터 2 | 데이터 3 |
| 데이터 4 | 데이터 5 | 데이터 6 |
- Json : key = value
    <Br>ex)
    - 역할 = 강사
    - 나이 = 26세 
- Symbol
    - 특수문자등을 이용해서 전달하는 프롬프트의 중요 부분을 강조
    - +, -, :, #, {}, """~"""
<br> -> 데이터를 구조화해서 문서를 작성하는 방법

```python
from openai import OpenAI
client = OpenAI(api_key=my_variable)

system = """
방법1
"다음의 합을 알려줘. 1, 2, 3, 4, 5, 6"

방법2
나는 너한테 리스트를 전달할거야
리스트의 합을 알려줘
List = [1, 2, 3, 4, 5, 6]
"""

user = """
방법1
"다음의 합을 알려줘. 1, 2, 3, 4, 5, 6"

방법2
나는 너한테 리스트를 전달할거야
리스트의 합을 알려줘
List = [1, 2, 3, 4, 5, 6]
"""

completion = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": system},
        {"role": "user", "content": user}
    ]
)

print(completion.choices[0].message.content)
```
